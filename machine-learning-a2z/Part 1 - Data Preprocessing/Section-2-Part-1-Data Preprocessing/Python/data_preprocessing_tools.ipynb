Independent Variables: These are the features or input variables that are used to make a prediction. In this case, the model will use the country, age, and salary of a customer to try and predict their purchase behavior.

Dependent Variable: This is the target or output variable that the model is trying to predict. It is called "dependent" because its value is expected to depend on the independent variables. In this dataset, 'Purchased' is the dependent variable, which likely has a binary outcome (e.g., Yes/No)

here in this example..

Country, age and salary are independent variables 
purchased is a dependent variable 

```python
# Selects the the first three columns (except the last one) / features - country, age, salary
X = dataset.iloc[:, :-1].values

# select the last column 
y = dataset.iloc[:, -1].values
print(X)
```

```python
print(y)
```

## Taking care of missing data


```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

# typically you use this method to select and apply this to all the numerical columns
# fit method will look at the missing valkues of the salary column and will calculate the average salary 
imputer.fit(X[:, 1:3])

# transform method will apply the changes to those rows
X[:, 1:3] = imputer.transform(X[:, 1:3])
```

```python
print(X)
```

## Encoding categorical data

In this example we have three distinct 

### Encoding the Independent Variable

```python
print (X)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# What happens below in the ColumnTransformer
# kind of transformation = encoding, what kind of encoding = one hot encoding
# third is the index of the column we want to encode [0] i.e. the country column
# remainder = 'passthrough' (leaves the other columns out of encoding)
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
# we need the output to be a numpy array as the train function further down the line will expect that
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

### Encoding the Dependent Variable (Y-variable)

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

## Splitting the dataset into the Training set and Test set

Here we split the dataset into two - training set and the test set 

What the split 
the training set is used to train the machine learning model  
the test set is used to evaluate the performance of the model based on these new observations

Then apply the feature scaling 
Why ? Becasue applying the scaling to the entire data set will consider the means / averages of the entire population this will skew the test set values i.e. it will induce information leakage

So ..to avoid this .. split the dataset first and then apply the feature scaling 

What should the split ratio be?
Recommended 80:20 



```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```
